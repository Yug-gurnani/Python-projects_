```python
def display_board(board):
    print(board[7]+'|'+board[8]+'|'+board[9])
    print('------')
    print(board[4]+'|'+board[5]+'|'+board[6])
    print('------')
    print(board[1]+'|'+board[2]+'|'+board[3])
    
test_board=['#','X','X','O','O','X','X','X','O','X']
    
def player_input():
    i = True
    while i == True:
        m = input()
        if m == 'X' or m=='O' or m=='x' or m=='o':
            i=False
            print('Enter the marker you want to choose: {}'.format(m))
        else:
            print('wrong input')
            break
    return m
            
def place_marker(board,marker,position):
    board[position]=marker
    display_board(board)




def win_check(board):
    return board[7]==board[8]==board[9]!=' ' or board[1]==board[2]==board[3]!=' ' or board[4]==board[5]==board[6]!=' '  or board[7]==board[4]==board[1]!=' ' or board[9]==board[6]==board[3]!=' ' or board[8]==board[5]==board[2]!=' ' or board[9]==board[5]==board[1]!=' ' or board[7]==board[5]==board[3]!=' '
    

    
import random

def choose_first():
    c = random.randint(1,2)
    print('player {} will go first'.format(c))
    return c
    
def space_check(board,position):
    return board[position]==' '

def full_board_check(board):
    for i in range(1,len(board)):
        if board[i]==' ':
            return False
    else:
        return True
    
def player_choice(board):
    print('Make the next move:')
    q = int(input())
    space_check(board,q)
    
    if space_check and q in range(1,10):
        return q
    else:
        return 'invalid number'

    
def replay():
    print('do you want to play again: yes/no')
    w = input()
    if w =='yes':
        return True
    else:
        return False
    
    

            
            
    
```

```python
def driver_code():
    board = ['#',' ',' ',' ',' ',' ',' ',' ',' ',' ']
    print('welcome to tic tac toe!')
    print('Enter the Marker you want to choose:')
    p1=player_input()
    if p1 == 'X' or p1=='x':
        p2= 'O'
    else:
        p2= 'X'
    current_chance=choose_first()
    if current_chance==1:
        current_chance=p1
    else:
        current_chance=p2
    
    while not win_check(board):
        if current_chance==p1:
            mark = player_choice(board)
            place_marker(board,p1,mark)
            current_chance=p2
        
            if win_check(board):
                print('player 1 won')
                break
        
        if current_chance==p2:
            mark = player_choice(board)
            place_marker(board,p2,mark)
            current_chance=p1
        
            if win_check(board):
                print('player 2 won')
                break
        
        
        if full_board_check(board):
            print('Game is Draw')
            break
            
```

```python
driver_code()
while replay():
    
    driver_code()
    
else:
    print('Game Over')
```
